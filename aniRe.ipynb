````
import shutil
data_raw = '/content/drive/My Drive/AnimalsRecognition/animals/raw-img'
base_dir='/content/drive/My Drive/AnimalsRecognition/animals'
lst_class = os.listdir(data_raw)


train_dir = os.path.join(base_dir, 'train')
test_dir = os.path.join(base_dir, 'test')
frame = {}
for class_animal in lst_class:
    temp = os.path.join(data_raw, class_animal)
    frame[class_animal] = os.listdir(temp)

    try:
        os.mkdir(train_dir +"/"+ class_animal)
    except:
        print('Cannot created train dir')

    try:
        os.mkdir(test_dir +"/"+ class_animal)
    except:
        print('Cannot created validation dir')
````

````
for animal_class in lst_class:
    rawfile_dir = os.path.join(data_raw, animal_class)
    train = int(len(frame[animal_class])*0.8)
    tem = os.path.join(train_dir, animal_class)
    for x in frame[animal_class][:train]:
        cur_dir = os.path.join(rawfile_dir, x)
        shutil.move(cur_dir, tem)
    tem = os.path.join(test_dir, animal_class)
    for x in frame[animal_class][train:]:
        cur_dir = os.path.join(rawfile_dir, x)
        shutil.move(cur_dir, tem)
````

````
train_dir='/content/drive/My Drive/AnimalsRecognition/animals/train'
test_dir='/content/drive/My Drive/AnimalsRecognition/animals/test'

def load_dataset(path):
    data = load_files(path)
    files = np.array(data['filenames'])
    targets = np.array(data['target'])
    target_labels = np.array(data['target_names'])
    return files,targets,target_labels

x_train, y_train,target_labels = load_dataset(train_dir)
x_test, y_test,_ = load_dataset(test_dir)
print('Loading complete!')

print('Training set size : ' , x_train.shape[0])
print('Testing set size : ', x_test.shape[0])
````
